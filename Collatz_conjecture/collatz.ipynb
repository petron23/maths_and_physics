```python
import matplotlib.pyplot as plt

def collatz_sequence(n):
    sequence = [n]
    while n != 1:
        if n % 2 == 0:
            n //= 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence

def collatz_sequence2(n):
    sequence = [n]
    for i in range(n,2000):
        if n % 2 == 0:
            n //= 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence

def plot_collatz(sequence):
    x_values = list(range(len(sequence)))
    plt.plot(x_values, sequence, marker='o', linestyle='-')
    plt.xlabel('Step')
    plt.ylabel('Value')
    plt.title('Collatz Conjecture Sequence')
    #plt.show()

# Example: Generate and plot the Collatz sequence
starting_number = 1
collatz_seq = collatz_sequence2(starting_number)
plot_collatz(collatz_seq)
```

```python
lenseq=[]
val = []
val2 = []
val3=[]
val4=[]

for i in range(1,2000):
    starting_number = i
    collatz_seq = collatz_sequence2(starting_number)
    plot_collatz(collatz_seq)
    lenseq.append(len(collatz_seq))
    val.append(collatz_seq[1])
    val2.append(collatz_seq[2])
    val3.append(collatz_seq[3])
    val4.append(collatz_seq[4])
    
plt.show()
```
