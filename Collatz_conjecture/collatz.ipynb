```python
import matplotlib.pyplot as plt

def collatz_sequence(n):
    sequence = [n]
    while n != 1:
        if n % 2 == 0:
            n //= 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence

def plot_collatz(sequence):
    x_values = list(range(len(sequence)))
    plt.plot(x_values, sequence, marker='o', linestyle='-')
    plt.xlabel('Step')
    plt.ylabel('Value')
    plt.title('Collatz Conjecture Sequence')
    #plt.show()

# Example: Generate and plot the Collatz sequence
starting_number = 4
collatz_seq = collatz_sequence(starting_number)
plot_collatz(collatz_seq)
```

```python
for i in range(1,47):
    starting_number = i
    collatz_seq = collatz_sequence(starting_number)
    plot_collatz(collatz_seq)
plt.show()
```
